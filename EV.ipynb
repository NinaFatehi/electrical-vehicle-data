```python
#In order to fill in the missing values in Vehicle Location column:
#find a the most common Vehicle Location's in each city and fill in the missing Vehicle Location with that.
for i in df.index:
  if pd.isna(df.loc[i, "Vehicle Location"]):
    if not pd.isna(df.loc[i, "City"]):
        city = df.loc[i, "City"]
        mode_value = df[df["City"]==city]["Vehicle Location"].mode()
        if not mode_value.empty:
          mode_value = mode_value[0]
          df.loc[i, "Vehicle Location"] = mode_value
        else:
          df.loc[i, "Vehicle Location"] = "Unknown"

```

```python
# 9 Vehcile locations are "Unknown"
len(df[df["Vehicle Location"]=="Unknown"])
```

```python
# To handle the missing values in the Electric Utility, I am going to replace the utility with the name of the city!
```

```python
# Relationship between Electric Utility and City
utility_by_city = df.groupby(['City', 'Electric Utility'])['Electric Utility'].count()
```

```python
utility_by_city
```

```python
utility_by_city = utility_by_city.rename("UtilityCount")
utility_by_city = utility_by_city.reset_index()
utility_by_city
```

```python
city_utility_dict = {}
for index, row in utility_by_city.iterrows():
  city = row['City']
  utility = row['Electric Utility']
  if city not in city_utility_dict:
    city_utility_dict[city] = utility
```
